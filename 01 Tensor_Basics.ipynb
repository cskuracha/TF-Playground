#### band_part

https://www.tensorflow.org/api_docs/python/tf/linalg/band_part

```
tf.linalg.band_part(
    input: Annotated[Any, TV_MatrixBandPart_T],
    num_lower: Annotated[Any, TV_MatrixBandPart_Tindex],
    num_upper: Annotated[Any, TV_MatrixBandPart_Tindex],
    name=None
) -> Annotated[Any, TV_MatrixBandPart_T]

```

***
<span style = 'color:green;'>This deals with lower and upper triangle along with diagonal matrices. Below are the various behaviour:

<span style = 'color:red;'>
    
1. When lower and upper is 0 -> Return the diagonal matrix
2. When lower and upper are -1, then entire matrix is replicated
3. When lower is 0 and upper is -1, this returns diagonal and upper matrix
4. When lower is -1 and upper is 0, this returns diagonal and lower matrix
5. When lower is 1 and upper is 1, then main diagonal and first subdiagonals are restoreed and rest will be zero
6. When lower is 1 and upper is 2, then main diagonal, first lower subdiagonal and 2 upper subdiagonals are restored, rest will be zero
